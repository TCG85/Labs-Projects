**Regression SalePrice - LotArea:**
- We see that an increase of 1 in LotArea causes a price increase of 2.1

```python
#assign constant

df_clean = sm.add_constant(df_clean)

y = df_clean["SalePrice"]
X = df_clean[["const", "LotArea"]]

quadr_reg = sm.OLS(y, X)
results_quadr = quadr_reg.fit()
results_quadr.summary()
```

**If we check for the influence of 
"YearRemodAdd", "OverallCond", "OverallQual"
on the SalePrice we see that:**
1. Every year that the refurbishment of the house is less away, it will increase the price by 407 (under the assumption that all other variable stay stable)
2. If the overall condition increase by 1, the price will drop by 1318 (this sounds counter intuitive, most likely it is related to the fact that most of the houses have a condition of 5 and that sales prices of hosues with a higher condition are not sold for more)
3. Opposite to the condition, the quality seems to be the characteristic with which the value of a house is significantly influenced - an increase by 1 in overall quality will increase the price by 41990

```python
y = df_clean["SalePrice"]
X = df_clean[["const", "YearRemodAdd", "OverallCond", "OverallQual"]]

quadr_reg = sm.OLS(y, X)
results_quadr = quadr_reg.fit()
results_quadr.summary()
```

**Influence of YearRemodAdd and YearBuilt:**
- A younger house by 1 year increases the price by 901 whereas a 1 year younger refurbishment increases the price by 1169.

```python
y = df_clean["SalePrice"]
X = df_clean[["const", "YearRemodAdd", "YearBuilt"]]

quadr_reg = sm.OLS(y, X)
results_quadr = quadr_reg.fit()
results_quadr.summary()
```

**Influence of Year/Month sold on sales price:**
- The later a house is sold during the year, the higher the price will be. If we remember that the majority of houses was sold during summer, there will be less houses available during winter, hence the drecreased supply will increase the price
- As in younger years the sales price decreases by 1354 per year, this might be an effect of the financial crisis in 2008/2009 which brought a lot of houses for sale on the market for quite low prices

```python
y = df_clean["SalePrice"]
X = df_clean[["const", "MoSold", "YrSold"]]

quadr_reg = sm.OLS(y, X)
results_quadr = quadr_reg.fit()
results_quadr.summary()
```
