```python
import pandas as pd
```

```python
#import file
df = pd.read_csv("Fortune 500 Companies US.csv", low_memory=False, encoding='latin-1')
```

```python
#show head
df.head(5)
```

```python
df.dtypes
```

```python
#remove commas
df["Number of Employees"] = df["Number of Employees"].str.replace(",", "")
df["Revenues"] = df["Revenues"].str.replace(",", "")
df["Profits"] = df["Profits"].str.replace(",", "")
df["Assets"] = df["Assets"].str.replace(",", "")
df["Market Value"] = df["Market Value"].str.replace(",", "")

#remove '$'
df["Revenues"] = df["Revenues"].str.replace("$", "")
df["Profits"] = df["Profits"].str.replace("$", "")
df["Assets"] = df["Assets"].str.replace("$", "")
df["Market Value"] = df["Market Value"].str.replace("$", "")
```

```python
#rename column names
(df.rename(columns={'Revenues': 'Revenues ($m)', 
                    'Profits': 'Profits ($m)', 
                    'Assets': 'Assets ($m)', 
                    'Market Value': 'Market Value ($m)'}, 
                    inplace=True))
```

```python
df.head(5)
```

```python
df.to_csv('Fortune_500_Companies_US_Clean.csv', sep=',', index=False)
```
