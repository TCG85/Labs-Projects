```python
import pandas as pd
```

```python
#import file
df = pd.read_csv("fortune1000.csv", low_memory=False, encoding='latin-1')
```

```python
#show head
df.head(5)
```

```python
#remove commas
df["Revenues ($M)"] = df["Revenues ($M)"].str.replace(",", "")
df["Profits ($M)"] = df["Profits ($M)"].str.replace(",", "")
df["Assets ($M)"] = df["Assets ($M)"].str.replace(",", "")
df["Mkt Value as of 3/29/18 ($M)"] = df["Mkt Value as of 3/29/18 ($M)"].str.replace(",", "")
df["Employees"] = df["Employees"].str.replace(",", "")

#remove '$'
df["Revenues ($M)"] = df["Revenues ($M)"].str.replace("$", "")
df["Profits ($M)"] = df["Profits ($M)"].str.replace("$", "")
df["Assets ($M)"] = df["Assets ($M)"].str.replace("$", "")
df["Mkt Value as of 3/29/18 ($M)"] = df["Mkt Value as of 3/29/18 ($M)"].str.replace("$", "")
```

```python
df.head(5)
```

```python
df.to_csv('fortune1000_clean.csv', sep=',', index=False)
```
